```python
# Try this approach
import requests
import pandas as pd



def read_google_sheet_csv(document_id, sheet_name):
    try:
        # Construct the URL
        url = f"https://docs.google.com/spreadsheets/d/{document_id}/gviz/tq?tqx=out:csv&sheet=UserDataEntry"
        
        # Use requests to fetch the content
        response = requests.get(url)
        
        # Check if the response is successful
        if response.status_code == 200:
            # Use StringIO to read the CSV content
            from io import StringIO
            df = pd.read_csv(StringIO(response.text))
            
            # Select the desired rows and columns
            forest_mgmt = df.iloc[7:12, 6:18]
            
            # Rename columns if needed
            # forest_mgmt.columns = ['Variable', 'Value_10', 'Value_20', 'Value_50']
            
            return forest_mgmt
        else:
            print(f"Failed to fetch the sheet. Status code: {response.status_code}")
            return None
    
    except Exception as e:
        print(f"Error reading the sheet: {e}")
        return None

# In your submit_form function
document_id = "1BRjiwdrDe0Q1vLFywMZwUALwNMOBYAPLrS_CWeiFilc"
forest_mgmt = read_google_sheet_csv(document_id, "ForestMgmtandHWPResults")
```
